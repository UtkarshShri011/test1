```python
import whylogs as why
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#dataframe
df = pd.read_csv("gmx_features_trades.csv")
results = why.log(df)
```

```python
df_train, df_test = train_test_split(df, test_size=0.2)
```

```python
results_test = why.log(df_test)
results_train = why.log(df_train)
```

```python
profile_train = results_train.profile()
profile_test = results_test.profile()

prof_train_view = profile_train.view()
prof_test_view = profile_train.view()
df.columns
```

```python
import whylogs as why
from whylogs.core.constraints import Constraints, ConstraintsBuilder
from whylogs.core.constraints.factories import greater_than_number,is_in_range,no_missing_values,column_is_nullable_datatype
from whylogs.core.constraints.factories import mean_between_range,stddev_between_range,quantile_between_range

```

```python
def is_mean_in_range_of_refrence_df_iqr(refrence_df,column,builder):
    Q1 = refrence_df[column].quantile(0.25)
    Q3 = refrence_df[column].quantile(0.75)

    builder.add_constraint(mean_between_range(column_name=column,lower=Q1,upper=Q3))
```

```python
builder = ConstraintsBuilder(prof_test_view)
builder.add_constraint(greater_than_number(column_name="log_open_entry_funding_rate", number=0.15))
is_mean_in_range_of_refrence_df_iqr(df_train,column='log_open_entry_funding_rate',builder= builder)
#builder.add_constraint(is_in_range(column_name='long_open_interest',lower=64803907.02109865,upper=-12626908.451910913))

constraints = builder.build()
constraints.generate_constraints_report()
```

```python
visualization.constraints_report(constraints, cell_height=300)
```
